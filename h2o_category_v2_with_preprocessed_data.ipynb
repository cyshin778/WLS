## GBM

```python
import h2o
import matplotlib.pyplot as plt
import pandas as pd

# h2o 초기화
h2o.init()

target = 'Depression_score_R6'

variable_dict = {
    'covariate': covariate,
    'econstab': econstab,
    'work': work,
    'education': education,
    'socialcommunity': socialcommunity,
    'healthcare': healthcare,
    'behavior': behavior,
    'gps': gps,
    'personality': personality
}

categories = ['econstab', 'work', 'education', 'socialcommunity', 'healthcare', 'behavior', 'gps','personality']

results = []

# 1) baseline model: covariate만 사용
print("Training baseline model (covariate only)")
X_baseline = variable_dict['covariate']
train_hf_baseline = train[X_baseline + [target]]
test_hf_baseline = test[X_baseline + [target]]

model_baseline = H2OAutoML(max_runtime_secs = 600, include_algos=["GBM"], seed = 2023, nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
model_baseline.train(x=X_baseline, y=target, training_frame=train_hf_baseline)

perf_baseline = model_baseline.leader.model_performance(test_hf_baseline)
results.append({'category': 'baseline (covariate only)', 'rmse': perf_baseline.rmse(), 'r2': perf_baseline.r2()})


# 2) 각 카테고리별 + covariate 모델
for cat in categories:
    print(f"Training with covariate + {cat}")

    X = variable_dict['covariate'] + variable_dict[cat]

    train_hf = train[X + [target]]
    test_hf = test[X + [target]]

    model = H2OAutoML(max_runtime_secs = 600, include_algos=["GBM"], seed = 2023, nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
    model.train(x=X, y=target, training_frame=train_hf)

    perf = model.leader.model_performance(test_hf)
    rmse = perf.rmse()
    r2 = perf.r2()

    results.append({'category': f'covariate + {cat}', 'rmse': rmse, 'r2': r2})

# 3) full model: covariate + 모든 카테고리 변수 합침
print("Training full model (covariate + all categories)")
X_full = variable_dict['covariate'] + sum([variable_dict[cat] for cat in categories], [])
train_hf_full = train[X_full + [target]]
test_hf_full = test[X_full + [target]]

model_full = H2OAutoML(max_runtime_secs = 600, include_algos=["GBM"], seed = 2023, nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
model_full.train(x=X_full, y=target, training_frame=train_hf_full)

perf_full = model_full.leader.model_performance(test_hf_full)
results.append({'category': 'full model (covariate + all)', 'rmse': perf_full.rmse(), 'r2': perf_full.r2()})

results_df = pd.DataFrame(results)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 정렬
results_df = results_df.sort_values('r2', ascending=True)

# 색상 생성
cmap = cm.get_cmap('tab20c', len(results_df))
colors = [cmap(i) for i in range(len(results_df)+1)]

# 그래프
plt.figure(figsize=(8, 6))
bars = plt.barh(results_df['category'],results_df['r2'], color=colors)

# 값 라벨 추가
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.005, bar.get_y() + bar.get_height()/2, f'{width:.3f}', va='center')

# 기준선
baseline_r2 = results_df[results_df['category'] == 'baseline (covariate only)']['r2'].values[0]
full_r2 = results_df[results_df['category'] == 'full model (covariate + all)']['r2'].values[0]

plt.axvline(x=baseline_r2, color='k', linestyle='--', label='Baseline')
plt.axvline(x=full_r2, color='k', linestyle=':', label='Full Model')

plt.xlabel('R^2 on Test Set')
plt.title(f'GBM, Target: {target}')

plt.legend()
plt.tight_layout()
plt.show()
```

## 개별 변수 예측력 (GLM)

```python
def get_wave(varname):
    if "_R3" in varname:
        return "R3"
    elif "_R4" in varname:
        return "R4"
    else:
        return "R5"

def get_valid_covariates(wave):
    if wave == "R3":
        return covariate_r3
    elif wave == "R4":
        return covariate_r3 + covariate_r4
    else:
        return covariate_r3 + covariate_r4 + covariate_r5

covariate_r3 = ['age_R3', 'sex', 'IQ', 'EV1', 'EV2', 'EV3', 'EV4', 'EV5', 'EV6', 'EV7', 'EV8', 'EV9', 'EV10']

covariate_r4 = ['age_R4', 'Alcohol_childhood_R4', 'Alcohol_live_R4', 'BMI_R4']

covariate_r5 = ['age_R5', 'Alcohol_childhood_R5', 'Alcohol_live_R5', 'Smoke_childhood_R5', 'BadExperience_R5', 'Childhood_ELS_R5', 'BMI_R5']

results_predictors = []

all_vars = econstab + work + education + socialcommunity + healthcare + behavior + gps + personality

target = 'Depression_score_R6'

for var in all_vars:
    wave = get_wave(var)
    X = get_valid_covariates(wave) + [var]

    train_hf = train[X + [target]]
    test_hf = test[X + [target]]

    # 모델 훈련 및 성능 평가
    glm_predictor_model = H2OAutoML(max_runtime_secs = 600, include_algos=["GLM"], seed = 2023, nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
    glm_predictor_model.train(x=X, y=target, training_frame=train_hf)
    perf = glm_predictor_model.leader.model_performance(test_hf)

    results_predictors.append({
            'variable': var,
            'rmse': perf.rmse(),
            'r2': perf.r2()})
```

```python
results_predictors = pd.DataFrame(results_predictors)
results_predictors = results_predictors.sort_values(by='r2', ascending=False).reset_index(drop=True)
results_predictors
```

```python
# 각 변수별 카테고리 매핑 함수 정의
def get_category(var):
    if var in econstab:
        return 'econstab'
    elif var in work:
        return 'work'
    elif var in education:
        return 'education'
    elif var in socialcommunity:
        return 'socialcommunity'
    elif var in healthcare:
        return 'healthcare'
    elif var in behavior:
        return 'behavior'
    elif var in gps:
        return 'gps'
    elif var in personality:
        return 'personality'
    else:
        return 'covariate'

# 카테고리 컬럼 추가
results_predictors['category'] = results_predictors['variable'].apply(get_category)
```
