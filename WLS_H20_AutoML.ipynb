## GBM

```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd

h2o.init()

target = 'Depression_score_R6'

variable_dict = {
    'covariate': covariate,
    'econstab': econstab,
    'work': work,
    'education': education,
    'socialcommunity': socialcommunity,
    'healthcare': healthcare,
    'behavior': behavior,
    'gps': gps,
    'personality': personality
}

categories = ['econstab', 'work', 'education', 'socialcommunity', 'healthcare', 'behavior', 'gps','personality']

results = []

def compute_metrics(model, test_hf, X, category_name):
    perf = model.leader.model_performance(test_hf)
    rmse = perf.rmse()
    mse = perf.mse()
    mae = perf.mae()
    r2 = perf.r2()
    mrd = perf.mean_residual_deviance()
    
    # Adjusted R^2 계산
    n = test_hf.nrows
    k = len(X)
    adj_r2 = 1 - (1 - r2) * ((n - 1) / (n - k - 1))
    
    return {
        'Category': category_name,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R^2': r2,
        'Adjusted R^2': adj_r2,
        'Mean Residual Deviance': mrd
    }

# 1) baseline model: covariates only
print("Training baseline model (covariate only)")
X_baseline = variable_dict['covariate']
train_hf_baseline = train_hf[X_baseline + [target]]
test_hf_baseline = test_hf[X_baseline + [target]]

model_baseline = H2OAutoML(max_runtime_secs=600, include_algos=["GBM"], seed=2023,
                           nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
model_baseline.train(x=X_baseline, y=target, training_frame=train_hf_baseline)

results.append(compute_metrics(model_baseline, test_hf_baseline, X_baseline, 'Baseline (covariate only)'))

# 2) categories
for cat in categories:
    print(f"Training with covariate + {cat}")
    X = variable_dict['covariate'] + variable_dict[cat]

    train_hf = train_hf[X + [target]]
    test_hf = test_hf[X + [target]]

    model = H2OAutoML(max_runtime_secs=600, include_algos=["GBM"], seed=2023,
                      nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
    model.train(x=X, y=target, training_frame=train_hf)

    results.append(compute_metrics(model, test_hf, X, f'{cat}'))

# 3) full model
print("Training full model (covariate + all categories)")
X_full = variable_dict['covariate'] + sum([variable_dict[cat] for cat in categories], [])
train_hf_full = train_hf[X_full + [target]]
test_hf_full = test_hf[X_full + [target]]

model_full = H2OAutoML(max_runtime_secs=600, include_algos=["GBM"], seed=2023,
                       nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
model_full.train(x=X_full, y=target, training_frame=train_hf_full)

results.append(compute_metrics(model_full, test_hf_full, X_full, 'Full model'))

results = results.sort_values("Adjusted R^2", ascending=False)

results_df = pd.DataFrame(results)
results_df.to_csv("GBM_cat_comparison.csv", index=False)
print("All results saved")
```

## Individual predictor's predictability (GLM)

```python
def get_wave(varname):
    if "_R3" in varname:
        return "R3"
    elif "_R4" in varname:
        return "R4"
    elif "_R5" in varname:
        return "R5"
    else:
        return "GPS"

def get_valid_covariates(wave):
    if wave == "R3":
        return base + age_R3
    elif wave == "R4":
        return base + age_R4
    elif wave == "R5":
        return base + age_R5
    else:
        return base 

base = ['sex', 'EV1', 'EV2', 'EV3', 'EV4', 'EV5', 'EV6', 'EV7', 'EV8', 'EV9', 'EV10']

age_R3 = ['age_R3']

age_R4 = ['age_R4']

age_R5 = ['age_R5']

results_predictors = []

demographic_subset = ['IQ','age_R3','age_R4', 'Alcohol_childhood_R4', 'Alcohol_live_R4', 'BMI_R4',
                      'age_R5', 'Alcohol_childhood_R5', 'Alcohol_live_R5', 'Smoke_childhood_R5', 'BadExperience_R5', 'Childhood_ELS_R5', 'BMI_R5']
                      

all_vars = demographic_subset + econstab + work + education + socialcommunity + healthcare + behavior + gps + personality

target = 'Depression_score_R6'

for var in all_vars:
    wave = get_wave(var)
    X = get_valid_covariates(wave) + [var]

    train_hf = train[X + [target]]
    test_hf = test[X + [target]]

    glm_predictor_model = H2OAutoML(max_runtime_secs = 600, include_algos=["GLM"], seed = 2023, nfolds=5, stopping_metric="RMSE", sort_metric="RMSE")
    glm_predictor_model.train(x=X, y=target, training_frame=train_hf)
    perf = glm_predictor_model.leader.model_performance(test_hf)

    r2 = perf.r2()
    n = test_hf.nrows
    k = len(X)
    adj_r2 = 1 - (1 - r2) * ((n - 1) / (n - k - 1))
    
    results_predictors.append({
            'variable': var,
            'rmse': perf.rmse(),
            'mse' : perf.mse(),
            'mae' : perf.mae(),
            'r2': perf.r2(),
            'adj_r2' : adj_r2,
            'mrd' : perf.mean_residual_deviance()})
```

```python
results_predictors = pd.DataFrame(results_predictors)

def get_category(var):
    if var in econstab:
        return 'econstab'
    elif var in work:
        return 'work'
    elif var in education:
        return 'education'
    elif var in socialcommunity:
        return 'socialcommunity'
    elif var in healthcare:
        return 'healthcare'
    elif var in behavior:
        return 'behavior'
    elif var in gps:
        return 'gps'
    elif var in personality:
        return 'personality'
    else:
        return 'covariate'

results_predictors['category'] = results_predictors['variable'].apply(get_category)
print(results_predictors.dtypes)
```
